```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
size = 100 
radius = 40

x = np.linspace(-size / 2, size / 2, size)
y = np.linspace(-size / 2, size / 2, size)
X, Y = np.meshgrid(x, y)

distance = np.sqrt(X**2 + Y**2)

circle = np.clip(1 - distance / radius, 0, 1) 
```

`pcolormesh`:

```python
plt.pcolormesh(X, Y, circle)
```

```python
plt.pcolormesh(X, Y, circle, cmap="bwr", vmin=0.0, vmax=0.5)
```

```python
plt.pcolormesh(X, Y, circle, cmap="bwr", vmin=0.0, vmax=1.0)
plt.colorbar(label="valor")
```

`contourf`:

```python
plt.contourf(X, Y, circle)
```

```python
plt.contourf(X, Y, circle, levels=4)
```

```python
plt.contourf(X, Y, circle, levels=4, cmap="Greens", vmin=0, vmax=1)
plt.colorbar(label="valor")
```
