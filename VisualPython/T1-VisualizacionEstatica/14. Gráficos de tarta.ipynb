```python
import matplotlib.pyplot as plt
```

```python
labels = ['Cat 1', 'Cat 2', 'Cat 3', 'Cat 4']
sizes = [35, 20, 11, 18]
```

```python
plt.pie(sizes, labels=labels)
```

```python
plt.pie(sizes, labels=labels, radius=2)
```

```python
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
```

```python
plt.pie(sizes, labels=labels, autopct='%1.1f%%', pctdistance=1.25, labeldistance=.6)
```

```python
plt.pie(sizes, labels=labels, colors=["pink", "yellow", "purple", "blue"])
```

```python
plt.pie(sizes, labels=labels, explode=(0, 0.3, 0, 0.2))
```
