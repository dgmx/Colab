```python
import matplotlib.pyplot as plt
```

Interfaz orientada a objetos:

```python
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4, 5], [1, 4, 9, 16, 25])
ax.set_xlabel('x')  
ax.set_ylabel('y')
ax.set_title('Gráfico con la interfaz orientada a objetos')
```

Interfaz de `pyplot`:

```python
plt.figure()
plt.plot([1, 2, 3, 4, 5], [1, 4, 9, 16, 25])
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico con la interfaz de pyplot')
```
