```python
import matplotlib.pyplot as plt
import numpy as np
```

Gráfico de barras simple:

```python
plt.bar(["cat 1", "cat 2", "cat 3"], [100, 70, 95])
```

```python
plt.bar(["cat 1", "cat 2", "cat 3"], [100, 70, 95], width=0.5, bottom=10, color="green")
```

Líneas de error:

```python
plt.bar(["cat 1", "cat 2", "cat 3"], [100, 70, 95], yerr=[4, 5, 6])
```

Gráfico de barras horizontal:

```python
plt.barh(["cat 1", "cat 2", "cat 3"], [100, 70, 95], height=0.5, left=10, color="orange")
```

Gráfico de barras múltiple:

```python
bar_width = 0.25

x = np.array([1, 2, 3])

plt.bar(x, [100, 70, 95], width=0.25, color="red", label="Red")
plt.bar(x + bar_width, [80, 80, 70], width=0.25, color="green", label="Green")
plt.bar(x + 2*bar_width, [90, 90, 90], width=0.25, color="blue", label="Blue")

plt.xticks(x + bar_width, ["cat 1", "cat 2", "cat 3"])

plt.legend()
```

Gráfico de barras apilado:

```python
cats = ["cat 1", "cat 2", "cat 3"]

reds = np.array([100, 70, 95])
greens = np.array([20, 10, 15])
blues = np.array([90, 90, 90])

plt.bar(cats, reds, label="Red")
plt.bar(cats, greens, bottom=reds, label="Green")
plt.bar(cats, blues, bottom=reds+greens, label="Blues")

plt.legend()
```
