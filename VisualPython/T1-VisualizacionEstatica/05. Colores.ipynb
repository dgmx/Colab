```python
import matplotlib.pyplot as plt
import numpy as np
```

Colores por defecto:

```python
x = np.linspace(0, 2 * np.pi, 1000)

sin = np.sin(x)
cos = np.cos(x)

plt.plot(x, sin)
plt.plot(x, cos)
```

Definiendo colores para `plot`:

```python
x = np.linspace(0, 2 * np.pi, 1000)

sin = np.sin(x)
cos = np.cos(x)

plt.plot(x, sin, color='red')
plt.plot(x, cos, c='green')
```

Otras formas de definir colores ([ms informacin](https://matplotlib.org/stable/users/explain/colors/colors.html#colors-def)):

```python
x = np.linspace(0, 10, 10)

plt.plot(x, 0.5 * x, color=(0, 0, 1)) # rgb (de 0 a 1)
plt.plot(x, 1.0 * x, color=(0, 0, 1, 0.5)) # rgba (de 0 a 1)
plt.plot(x, 1.5 * x, color='#00FF00') # hex
plt.plot(x, 2.0 * x, color='0.7') # grises (0.0 -> negro, 1.0 -> blanco)
plt.plot(x, 2.5 * x, color='cyan') # 'blue', 'green', 'red', 'cyan', 'magenta', 'yellow', 'black', 'white'
plt.plot(x, 3.0 * x, color="k") # 'b', 'g', 'r', 'c', 'm', 'y', 'k', 'w'
```

[Colormaps](https://matplotlib.org/stable/users/explain/colors/colormaps.html):

```python
plt.rcParams["axes.prop_cycle"]
```

```python
plt.rcParams["axes.prop_cycle"] = plt.cycler("color", plt.cm.Set3.colors)

x = np.linspace(0, 10, 10)

plt.plot(x, 0.5 * x)
plt.plot(x, 1.0 * x)
plt.plot(x, 1.5 * x)
plt.plot(x, 2.0 * x)
```

```python
plt.rcParams["axes.prop_cycle"] = plt.cycler("color", ['#ff0000', '#00ff00', '#0000ff'])

x = np.linspace(0, 10, 10)

plt.plot(x, 0.5 * x)
plt.plot(x, 1.0 * x)
plt.plot(x, 1.5 * x)
plt.plot(x, 2.0 * x)
```

Colores con `scatter`:

```python
x = np.random.uniform(size=100)
y = x + np.random.normal(loc=0.1, scale=0.1, size=100)

plt.scatter(x, y, color='#a3204a')
```

```python

x = np.linspace(0, 1, 15)
y = np.linspace(0, 1, 15)
xx, yy = np.meshgrid(x, y)

plt.scatter(xx, yy, c=xx+yy, cmap='rainbow')
```
