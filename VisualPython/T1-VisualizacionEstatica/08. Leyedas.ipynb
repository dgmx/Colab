```python
import matplotlib.pyplot as plt
import numpy as np
```

AÃ±adiendo leyendas:

```python
x = np.linspace(0, 6*np.pi, 1000)

plt.plot(x, np.cos(x), label="cos")
plt.plot(x, np.sin(x), label="sin")

plt.legend()
```

PosiciÃ³n de la leyenda:

```python
x = np.linspace(0, 6*np.pi, 1000)

plt.plot(x, np.cos(x), label="cos")
plt.plot(x, np.sin(x), label="sin")

plt.legend(loc="upper right") # upper/lower left/center/right
```

Leyendas fuera de la figura:

```python
x = np.linspace(0, 6*np.pi, 1000)

fig = plt.figure(constrained_layout=True)

plt.plot(x, np.cos(x), label="cos")
plt.plot(x, np.sin(x), label="sin")

fig.legend(loc="outside upper center") # upper/lower left/center/right
```

`ncols`:

```python
x = np.linspace(0, 6*np.pi, 1000)

fig = plt.figure(constrained_layout=True)

plt.plot(x, np.cos(x), label="cos")
plt.plot(x, np.sin(x), label="sin")

fig.legend(loc="outside upper center", ncols=2) # upper/lower left/center/right
```

Diferentes tipos de grÃ¡ficos en una leyenda:

```python
x = np.linspace(0, 6*np.pi, 100)

plt.plot(x, np.cos(x), label="cos")
plt.scatter(x, np.sin(x), label="sin", color="green", marker="s")

plt.legend()
```

`fontsize`:

```python
x = np.linspace(0, 6*np.pi, 100)

plt.plot(x, np.cos(x), label="cos")
plt.scatter(x, np.sin(x), label="sin", color="green", marker="s")

plt.legend(fontsize=18, loc="upper right", ncols=2, framealpha=1) # por defecto framealpha=0.8
```
