```python
import matplotlib.pyplot as plt
import numpy as np
```

`subplots`:

```python
x = np.arange(100)

fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(16,9))

ax[0,0].plot(x, x)
ax[0,1].plot(x, x**2)
ax[0,2].plot(x, x**3)

ax[1,0].plot(x, -x)
ax[1,1].plot(x, -x**2)
ax[1,2].plot(x, -x**3)
```

```python
x = np.arange(100)

fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(16,4.5))

ax[0].plot(x, x)
ax[1].plot(x, x**2)
ax[2].plot(x, x**3)
```

```python
x = np.arange(100)

fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(16,9), sharex='all', sharey='row') # all, none, col, row

ax[0,0].plot(x, x)
ax[0,1].plot(x, x**2)
ax[0,2].plot(x, x**3)

ax[1,0].plot(x, -x)
ax[1,1].plot(x, -x**2)
ax[1,2].plot(x, -x**3)
```

`subplot_mosaic`:

```python
x = np.arange(100)

fig, ax = plt.subplot_mosaic(mosaic=[['a', 'a', 'b'], ['c', 'd', 'b']], figsize=(12,12), gridspec_kw={'wspace':0.2,'hspace':0.2})

ax['a'].plot(x, x)
ax['b'].plot(x, x**2)
ax['c'].plot(x, x**3)
ax['d'].plot(x, -x)
```
