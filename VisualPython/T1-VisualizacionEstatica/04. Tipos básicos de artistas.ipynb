```python
import matplotlib.pyplot as plt
import numpy as np
```

Generación de gráficos con `plot`:

```python
x = np.linspace(0, 2 * np.pi, 1000)
sin = np.sin(x)
cos = np.cos(x)
```

```python
plt.plot(x, sin)
plt.plot(x, cos)
```

Generación de gráficos con `scatter`:

```python
x = np.random.uniform(size=100)
y1 = x + np.random.normal(loc=0.1, scale=0.1, size=100)
y2 = 0.5 - x + np.random.normal(loc=0.1, scale=0.1, size=100)
```

```python
plt.scatter(x, y1)
plt.scatter(x, y2)
```

Generación de gráficos combinados:

```python
x1 = np.linspace(0, 1, 100)
y1 = -x1

x2 = np.random.uniform(size=100)
y2 = x2 + np.random.normal(loc=0.1, scale=0.1, size=100)
```

```python
plt.plot(x1, y1)
plt.scatter(x2, y2)
```
