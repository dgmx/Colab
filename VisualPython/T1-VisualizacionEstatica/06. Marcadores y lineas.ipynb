```python
import matplotlib.pyplot as plt
import numpy as np
```

`linewidth`:

```python
x = np.linspace(0, 10, 10)

plt.plot(x, 0.5 * x, linewidth=0.5)
plt.plot(x, 1.0 * x, linewidth=1.0) # valor por defecto
plt.plot(x, 1.5 * x, linewidth=1.5)
plt.plot(x, 2.0 * x, lw=2.0) # notaci√≥n alternativa
```

`linestyle`:

```python
x = np.linspace(0, 10, 10)

plt.plot(x, 0.50 * x, linestyle='-')
plt.plot(x, 0.75 * x, linestyle='solid')
plt.plot(x, 1.00 * x, linestyle='--')
plt.plot(x, 1.25 * x, linestyle='dashed')
plt.plot(x, 1.50 * x, linestyle='-.')
plt.plot(x, 1.75 * x, linestyle='dashdot')
plt.plot(x, 2.00 * x, ls=':') 
plt.plot(x, 2.25 * x, ls='dotted') 
```

```python
x = np.linspace(0, 10, 10)

plt.plot(x, 0.50 * x, linestyle=(0,(10, 10)))
plt.plot(x, 1.00 * x, linestyle=(10,(10, 10))) # primer valor: desplazamiento

plt.plot(x, 2.00 * x, linestyle=(0,(10, 10)))
plt.plot(x, 2.50 * x, linestyle=(0,(20, 10))) # segundo valor: tama√±o de linea y espacio
plt.plot(x, 3.00 * x, linestyle=(0,(10, 10, 30, 10))) # el segundo valor tiene que ser una tupla con un n√∫mero par de valores
```

`marker`:

```python
x = np.linspace(0, 10, 10)

plt.plot(x, 0.50 * x, marker='.')
plt.plot(x, 1.00 * x, marker='.', linestyle='')
plt.plot(x, 1.50 * x, marker='.', markersize='10', linestyle='none')
```

```python
x = np.linspace(0, 10, 10)

plt.plot(x, 0.50 * x, ls='', marker='.', markersize=5)
plt.plot(x, 0.75 * x, ls='', marker=',', markersize=5)
plt.plot(x, 1.00 * x, ls='', marker='o', markersize=5)
plt.plot(x, 1.25 * x, ls='', marker='s', markersize=5)
plt.plot(x, 1.50 * x, ls='', marker='^', markersize=5)
plt.plot(x, 1.75 * x, ls='', marker='v', markersize=5)
plt.plot(x, 2.00 * x, ls='', marker='<', markersize=5) 
plt.plot(x, 2.25 * x, ls='', marker='>', markersize=5) 
plt.plot(x, 2.50 * x, ls='', marker='+', ms=5)
plt.plot(x, 2.75 * x, ls='', marker='x', ms=5) 

# otros marcadores: https://matplotlib.org/stable/api/markers_api.html#module-matplotlib.markers
```

```python
x = np.random.uniform(size=100)
y1 = x + np.random.normal(loc=0.1, scale=0.1, size=100)
y2 = 0.5 - x + np.random.normal(loc=0.1, scale=0.1, size=100)
plt.scatter(x, y1, marker='x')
plt.scatter(x, y2, marker='s')
```

`markerfacecolor` y `makeredgecolor`:

```python
x = np.linspace(0, 10, 10)

plt.plot(x, 0.50 * x, marker='o', markersize=10, color='blue')
plt.plot(x, 1.00 * x, marker='o', markersize=10, color='blue', markerfacecolor='pink')
plt.plot(x, 1.50 * x, marker='o', markersize=10, color='blue', markerfacecolor='pink', markeredgecolor='yellow')
```

`fmt`:

```python
# fmt = '[marker][line][color]'

x = np.linspace(0, 10, 10)

plt.plot(x, 0.50 * x, 's-r')
plt.plot(x, 1.00 * x, '--g')
plt.plot(x, 1.50 * x, '^-.b')
plt.plot(x, 2.00 * x, 'b')
plt.plot(x, 2.50 * x, 'or')
plt.plot(x, 3.00 * x, '-g')
plt.plot(x, 3.50 * x, '--')
plt.plot(x, 4.00 * x, '^k')
```
