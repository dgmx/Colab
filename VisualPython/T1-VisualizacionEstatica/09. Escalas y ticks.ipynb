```python
import matplotlib.pyplot as plt
import numpy as np
```

Escalas:

```python
x = np.arange(100)
y = x**3

plt.yscale('log') # linear, log
plt.plot(x, y)
```

```python
x = np.arange(100)
y = x**3

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_yscale('log')
```

`Tick locators` y `formaters`:

```python
x = np.arange(100)
y = x**3

fig, ax = plt.subplots()
ax.set_xticks([10, 40, 60, 100]) # locators
ax.plot(x, y)
```

```python
x = np.arange(100)
y = x**3

fig, ax = plt.subplots()
ax.set_xticks([10, 40, 60, 100], ['diez', 'cuarenta', 'sesenta', 'cien']) # locators, formaters
ax.plot(x, y)
```

Estilo de los `ticks` mediante [`tick_parms`](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.tick_params.html#matplotlib.axes.Axes.tick_params):

```python
x = np.arange(100)
y = x**3

fig, ax = plt.subplots()
ax.tick_params(axis='both', which='major', direction='inout') # in, out, inout
ax.plot(x, y)
```

```python
x = np.arange(100)
y = x**3

fig, ax = plt.subplots()
ax.tick_params(axis='both', which='major', direction='inout', bottom=False, top=True, left=False, right=True, labelbottom=False, labeltop=True, labelleft=False, labelright=True) 
ax.plot(x, y)
```

```python
x = np.arange(100)
y = x**3

fig, ax = plt.subplots()
ax.tick_params(axis='both', which='major', direction='inout', length=10.0, width=5.0)
ax.plot(x, y)
```

```python
x = np.arange(100)
y = x**3

fig, ax = plt.subplots()
ax.tick_params(axis='both', which='major', direction='inout', length=10.0, width=5.0, pad=15.0)
ax.plot(x, y)
```

```python
x = np.arange(100)
y = x**3

fig, ax = plt.subplots()
ax.tick_params(axis='both', which='major', direction='inout', length=10.0, width=5.0, pad=15.0, labelsize=20)
ax.plot(x, y)
```

`ticks` secundarios:

```python
x = np.arange(100)
y = x**3

fig, ax = plt.subplots()
ax.minorticks_on() # activa ambos ejes
ax.tick_params(axis='y', which='minor', left=False) # desactiva el eje y
ax.plot(x, y)
```

[`grid`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.grid.html):

```python
x = np.arange(100)
y = x**3

fig, ax = plt.subplots()
ax.grid()
ax.plot(x, y)
```

```python
x = np.arange(100)
y = x**3

fig, ax = plt.subplots()
ax.grid(which='major', axis='x')
ax.plot(x, y)
```
