```python
import matplotlib.pyplot as plt
import numpy as np
```

`xlabel`, `ylabel` y `title`:

```python
x = np.linspace(0, 10, 10)

plt.xlabel('eje x')
plt.ylabel('eje y')
plt.title('title')

plt.plot(x, x)
```

```python
fig, ax = plt.subplots()

x = np.linspace(0, 10, 10)

ax.set_xlabel('eje x')
ax.set_ylabel('eje y')
ax.set_title('title')

ax.plot(x, x)
```

`text`:

```python
x = np.linspace(0, 10, 10)

plt.xlabel('eje x')
plt.ylabel('eje y')
plt.title('title')

plt.text(0, 0, 'mi texto en 0,0')
plt.text(7, 3, 'mi texto en 7,3')

plt.plot(x, x)
```

Tipografías:

```python
# más información: https://matplotlib.org/stable/api/text_api.html#matplotlib.text.Text

x = np.linspace(0, 10, 10)

plt.xlabel('eje x', fontsize=20)
plt.ylabel('eje y', fontfamily='serif', fontsize=15) # 'serif', 'sans-serif', 'cursive', 'fantasy', 'monospace'
plt.title('title', color='red')

plt.plot(x, x)
```

TeX:

```python
x = np.linspace(0, 10, 100)

plt.xlabel('x', fontsize=20)
plt.ylabel(r'$x^2$', fontsize=20)

plt.plot(x, x**2)
```
