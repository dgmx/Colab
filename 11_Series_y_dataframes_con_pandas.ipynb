## Series  en Pandas

pdseries.png

Las series son como los arrays en pandas, como un array o dimensión en numpy

```python
import pandas as pd
```

Definimos un pd.Series donde el primer array corresponde al nombre del jugador y el segundo corresponde a el índice con el que se encontrará

```python

psg_players = pd.Series(['Navas', 'Mbappe','Neymar','Messi'],
          index=[1,7,10,30]
          )

psg_players
```

Si no defines el segundo array (index) toma una lista predeterminada ordenada desde 0

```python
pd.Series(['Navas', 'Mbappe','Neymar','Messi'])
```

También podemos definir una pd.Series con un diccionario de py

```python
dicc = {1:'Navas',7:'Mbappe', 10:'Neymar',30:'Messi'}
pd.Series(dicc)
```

Se pueden usar índices en las Series

```python
psg_players[30]
```

O slicing

```python
psg_players[0:3]
```

## Dataframe

pdseries.png

Es como una matriz en numpy (arrays en estructura matricial)


Con el siguiente diccionario estaríamos creando las columnas

```python
# Creamos un segundo diccionario pero de dos dimensiones (un framework)
dicc2 = {'Jugador':['Navas','Mbappe','Neymar','Messi'],
 'Altura':[183.0,170.0,170.0,165.0],
 'Goles':[2,200,200,200]}
dicc2
```

Con ese diccionario creamos un dataframe y con pd.dataframe agregamos los índices (filas)

```python
df_players = pd.DataFrame(dicc2, index=[1,7,10,30])
df_players
```

Si no agregamos un index se crea uno predeterminado por py

```python
pd.DataFrame(dicc2)
```
