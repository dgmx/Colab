```python
import pandas as pd
```

Para leer archivos csv usamos pd.read_csv(path)

```python
pd.read_csv('/content/archivoclase11.csv', sep = ',', header = 0)
```

Si mi archivo csv no está delimitado por comas, podemos cambiar el separador,  sep = ' , '

```python
pd.read_csv('/content/archivoclase11.csv', sep = ',')
```

Si el archivo no tiene header podemos especificarlo con header = indice del titulo de las columnas

Si no tiene título se especifica None

```python
pd.read_csv('/content/archivoclase11.csv', header = None)
```

Podemos guarda la información como un dataframe en variables

```python
df_books = pd.read_csv('/content/archivoclase11.csv', sep = ',', header = 0)
```

```python
#obtenemos el nombre de las columnas

df_books.columns
```

Para cambiar el nombre de las columnas usamos 'names'

```python
df_books = pd.read_csv('/content/archivoclase11.csv', sep = ',', header = 0, names= ['Nombre', 'Autor', 'Calificación', 'Reviews', 'Price', 'Year', 'Genre'])
df_books
```
