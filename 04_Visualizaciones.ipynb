# Exploración visual de los datos

Existen dos vertientes en la estadística descriptiva:

* Bloque analítica:
Habla de las medidas de tendencia central y de las medidas de dispersión

* Bloque visual:
Utiliza data viz para entender de manera visual los conceptos

Sugerencia para conocer tipos de data viz: https://datavizproject.com/#



```
# Importamos librerías

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```
#Importamos el dataset 'iris' de seaborn

iris = sns.load_dataset('iris')
iris
```

Nota: el dataset 'iris' tiene información de tres especies de una planta llamada 'iris', las características principales para agruparlas son el ancho y largo de sus sépalos y sus pétalos

```
iris.info()
```

https://seaborn.pydata.org/generated/seaborn.scatterplot.html

```
# scatterplot o diagrama de dispersión

sns.set_style(style='dark')
sns.scatterplot(data=iris, x='sepal_length',y='petal_length',hue='species')
plt.title('Diagrama de dispersión agrupado por especies de iris')
plt.show()
```

http://seaborn.pydata.org/generated/seaborn.jointplot.html

```
# joint plot

sns.jointplot(data=iris,x='sepal_width',y='petal_width',hue='species')
```

```
# box plot

sns.boxplot(data=iris, x='species',y='petal_width')
```

https://seaborn.pydata.org/generated/seaborn.barplot.html

```
#Barplot
sns.barplot(x = 'species', y = 'petal_length', data = iris)
```
