# Agregar o eliminar datos con pandas de un dataframe

```python
#Importamos la librer√≠a

import pandas as pd
import numpy as np
```

```python
#Importamos nuestro dataframe

df_books = pd.read_csv('/content/archivoclase12.csv', sep=',',header=0)
df_books.head(5)
```

##Drop

### Drop Columns

* df_books: este es el nombre del DataFrame con el que se est√° trabajando.
drop(): este es el m√©todo que se utiliza para eliminar una columna del DataFrame.
* 'Genre': este es el nombre de la columna que se desea eliminar del DataFrame.
* axis=1: este par√°metro indica que se desea eliminar una columna, en lugar de una fila, del DataFrame. El valor de 1 indica que la columna se elimina a lo largo del eje horizontal.
* inplace=True: este par√°metro indica que el cambio debe realizarse directamente en el DataFrame original, en lugar de crear una copia del DataFrame.

```python
df_books.drop('Genre', axis=1, inplace=True)
```

```python
df_books.head(5)
```

Tambi√©n se puede hacer de esta forma

```python
df_books = df_books.drop('Year',axis=1)
```

```python
df_books.head(5)
```
