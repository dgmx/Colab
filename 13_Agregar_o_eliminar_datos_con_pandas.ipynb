# Agregar o eliminar datos con pandas de un dataframe

```python
#Importamos la librería

import pandas as pd
import numpy as np
```

```python
#Importamos nuestro dataframe

df_books = pd.read_csv('/content/archivoclase12.csv', sep=',',header=0)
df_books.head(5)
```

##Drop

### Drop Columns

* df_books: este es el nombre del DataFrame con el que se está trabajando.
drop(): este es el método que se utiliza para eliminar una columna del DataFrame.
* 'Genre': este es el nombre de la columna que se desea eliminar del DataFrame.
* axis=1: este parámetro indica que se desea eliminar una columna, en lugar de una fila, del DataFrame. El valor de 1 indica que la columna se elimina a lo largo del eje horizontal.
* inplace=True: este parámetro indica que el cambio debe realizarse directamente en el DataFrame original, en lugar de crear una copia del DataFrame.

```python
df_books.drop('Genre', axis=1, inplace=True)
```

```python
df_books.head(5)
```

También se puede hacer de esta forma

```python
df_books = df_books.drop('Year',axis=1)
```

```python
df_books.head(5)
```

Se puede hacer con la función del  de python

```python
del df_books['User Rating']
```

```python
df_books.head(5)
```

### Drop rows

Es muy similar a borrar filas con .drop( )

```python
df_books.drop(2,axis=0).head(5)
```

También puedes borrar una lista de filas

```python
df_books.drop([0,3,6], axis=0).head(5)
```

Usa los rangos de python

```python
df_books.drop(range(0,10), axis=0).head(5)
```
